### sklearn.model_selection.GridSearchCV

*class sklearn.model_selection.GridSearchCV(estimator, param_grid, *, scoring=None, n_jobs=None, refit=True, cv=None, verbose=0, pre_dispatch='2*n_jobs', error_score=nan, return_train_score=False)

```python
import pandas as pd 
df = pd.read_csv('./datasets/iris.csv',encoding='utf-8')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
label = le.fit_transform(df['Species'])
df['label']=label
```

```python
df.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
X=df.iloc[:,1:5]
y=df.iloc[:,6]
y
```

```python
dt_clf = DecisionTreeClassifier(random_state=112)  

# X=iris[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
param_gird = {
    'max_depth':[3,5,7], # 분석가 마음 
    'min_samples_split':[3,5]
} ###### 이거 dictionary 로 fix 하자
clf = GridSearchCV(estimator =dt_clf,param_grid= param_gird,cv=5, scoring='accuracy', refit=True )
# refit = 최적의 하이퍼 파라미터 찾은 걸 estimator(모델)로 만들어서 특정 attribute에 저장 
clf.fit(X,y)
# cv해가지고 최적의 하이퍼 파라미터로 만들어 놓는다 
```

```python
clf.cv_results_
```

```python
clf.best_estimator_
```

```python
best_clf = clf.best_estimator_
```
